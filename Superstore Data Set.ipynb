```python
def outlier(x,fig):
    plt.subplot(2,2,fig)
    plt.title(x)
    sns.boxplot(x=data[x],color="red")
plt.figure(figsize=(20,10))
for e, i in enumerate(list1):
    outlier(i,e+1)
plt.show()
```
